```python
#function to add data labels
def text_clean(text):
    text=str(text).strip()
    return text

bat_data['Batsman']=bat_data['Batsman'].apply(text_clean)
```

### We have match record from 11-09-1999 to 19-05-2019

```python
#function to add data label
def add_label(x_list,y_list):
    for i in range(len(x_list)):
        plt.annotate(y_list[i],(i,y_list[i]), ha='center', fontsize=15)
```

<h2 style="color:blue; background:lightgreen; line-height:1.3">2. Top 10 Players with highest matches played between 1999 to 2019..</h2>

```python
plt.figure(figsize=(18,6))
sns.countplot(x='Batsman', data=bat_data,
             order=bat_data['Batsman'].value_counts().index[:10])
plt.xticks(fontsize=12)
plt.yticks(fontsize=15)
plt.title("Highest Matched Played by Player between 1999 to 2019", fontsize=18, color='blue')

index=bat_data['Batsman'].value_counts().index[:10]
value=bat_data['Batsman'].value_counts().values[:10]
add_label(index, value);
```

### MS Dhoni Played most matches between 1999 to 2019 period.

<h2 style="color:blue; background:lightgreen; line-height:1.3">3. Top 10 Players with higest runs scored between 1999 to 2019..</h2>

```python
plt.figure(figsize=(18,6))
sns.barplot(x='Batsman', y='Runs', data=bat_data, ci=0, estimator=sum,
             order=bat_data.groupby('Batsman').Runs.sum().sort_values(ascending=False).index[:10])
plt.xticks(fontsize=12)
plt.yticks(fontsize=15)
plt.title("Highest Runs by Player between 1999 to 2019", fontsize=18, color='blue')

index=bat_data.groupby('Batsman').Runs.sum().sort_values(ascending=False).index[:10]
value=bat_data.groupby('Batsman').Runs.sum().sort_values(ascending=False).values[:10]
add_label(index, value);
```
